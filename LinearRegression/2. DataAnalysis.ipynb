# Making subset from other dataset

---
# 'apply(func)' - applying the 'func' to row/column

You can choose what to apply it for:
- axis=0 -> columns (default); 
- axis=1 -> rows

```python
data_subset.apply(max, axis=0) 
```

---
# Using lambda is fine (Duh ???)

```python
data_subset.apply(lambda row: row*2, axis=1) 
```

---
# 'where(condition, value)' - conditional formatting (by each row)

Keeps all rows that vaildate the condition, rest with be changed into 'np.nan'

```python
data_subset = data_subset.where(data_subset.TAX > 250, np.nan) 
```

---
# 'map' - conditional formatting (by each cell)

```python
data_subset.map(lambda x: round(x) if type(x)!=type(np.nan) else x) 
```

---
 # Fills 'NaN' values with '-1'

```python
data_subset.fillna(-1)
```

# Plotting the data set visually

**kind : str**

- â€˜lineâ€™ : line plot (default)
- â€˜barâ€™ : vertical bar plot
- â€˜barhâ€™ : horizontal bar plot
- â€˜histâ€™ : histogram
- â€˜boxâ€™ : boxplot
- â€˜kdeâ€™ : Kernel Density Estimation plot
- â€˜densityâ€™ : same as â€˜kdeâ€™
- â€˜areaâ€™ : area plot
- â€˜pieâ€™ : pie plot
- â€˜scatterâ€™ : scatter plot
- â€˜hexbinâ€™ : hexbin plot

```python
data_plot = data[:5].plot(kind="bar")
plt.title("Boston Housing")
plt.style.use("dark_background")
data_plot.set_xlabel("Features")
data_plot.set_ylabel("Count")
plt.show()
```

---
# Plotting Correlation Matrix

sns.heatmap(annot=True) - Display exact correlation values

```python
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Bostong Housing Matrix")
plt.show()
```

# Pairplotting with Seaborn

```python
selected_features = ['RM', 'ZN', 'DIS', 'B', 'LSTAT', 'MEDV']
sns.pairplot(data[selected_features])
plt.suptitle("Boston Pairplot")
plt.show()
```
